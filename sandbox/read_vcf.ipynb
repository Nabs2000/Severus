This notebook reads in a VCF file as a Pandas DataFrame and performs some basic data exploration.

```python
# Import libraries
import pandas as pd
```

```python
# Read the VCF file
def read_vcf(file_path):
    with open(file_path, 'r') as file:
        lines = [line for line in file if not line.startswith('##')]
    # Extracting the header
    header_line = [line for line in lines if line.startswith('#')][0]
    header = header_line.strip('#').strip().split('\t')
    # Creating a DataFrame with the VCF data
    data = [line.strip().split('\t') for line in lines if not line.startswith('#')]
    return pd.DataFrame(data, columns=header)
```

```python
# File path
vcf_file_path = '../severus_out/somatic_SVs/severus_somatic.vcf'

# Load and parse the VCF file
vcf_df = read_vcf(vcf_file_path)
vcf_df.head(100)
```
